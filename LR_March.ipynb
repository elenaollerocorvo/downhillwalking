```python
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib.patches as mpatches
import matplotlib.patheffects as pe
import numpy as np
from scipy.interpolate import interp1d
from scipy import stats

# ============================================================
# PEAK DETECTION LOGIC — max or min per variable per band
# ============================================================
# 'max' = find the highest point in the band (default)
# 'min' = find the lowest point (e.g. GRF P2 in downhill)

peak_logic = {
    #  variable       P1 band    P2 band
    'GRF_vert':   ('max',      'max'),   # P1 ↑ with slope, P2 ↓ with slope
                                          # but each curve's own peak is still max within band
    'kneeFres':   ('max',      'max'),
    'knee_PFc':   ('max',      'max'),
    'angleKnee':  ('max',      'max'),
    'momentKnee': ('max',      'max'),
}

# For the trend annotation we compare the peak value at 0° vs -18°
# If value decreases → annotate ↓, if increases → annotate ↑

# ============================================================
# CONFIGURATION (same as before)
# ============================================================
variables_to_plot = {
    'GRF_vert':   'Vertical Ground Reaction Force (×BW)',
    'kneeFres':   'Tibiofemoral Joint Compression Force (×BW)',
    'knee_PFc':   'Patellofemoral Joint Compression Force (×BW)',
    'angleKnee':  'Knee Flexion Angle (°)',
    'momentKnee': 'Knee Extension Moment (Nm)',
}
# Añade esto antes del for loop (línea ~74)
df_plot = df_p1.copy()  # o el dataframe que quieras visualizar
slopes_ordered = [0, -6, -12, -18]
slope_labels   = {0: '0°', -6: '−6°', -12: '−12°', -18: '−18°'}

colors_slope = {
    0:   '#a8d5e2',
   -6:   '#5b9fbf',
  -12:   '#2a6a96',
  -18:   '#0d2d55',
}
linestyles = {
    0:   (0, ()),
   -6:   (0, (6, 2)),
  -12:   (0, (3, 1, 1, 1)),
  -18:   (0, (1, 1)),
}

stance_pct      = np.linspace(0, 100, 101)
absorption_band = (10, 30)
propulsion_band = (65, 85)

def stance_interp(arr, stance_fraction):
    arr = np.array(arr, dtype=float)
    n = len(arr)
    if n < 2:
        return np.full(101, np.nan)
    n_stance = max(2, round(stance_fraction * n))
    arr_s = arr[:n_stance]
    x_orig = np.linspace(0, 100, len(arr_s))
    f = interp1d(x_orig, arr_s, kind='linear',
                 bounds_error=False, fill_value=np.nan)
    return f(stance_pct)

# ============================================================
# BUILD CURVES
# ============================================================
mean_curves = {var: {} for var in variables_to_plot}
std_curves  = {var: {} for var in variables_to_plot}

for slope in slopes_ordered:
    df_s = df_plot[df_plot['slope'] == slope]
    for var in variables_to_plot:
        all_trials = []
        for _, row in df_s.iterrows():
            val = row[var]
            sf  = row['stance']
            if isinstance(val, (list, np.ndarray)) and len(val) > 1:
                all_trials.append(stance_interp(val, sf))
        if all_trials:
            stack = np.array(all_trials)
            mean_curves[var][slope] = np.nanmean(stack, axis=0)
            std_curves[var][slope]  = np.nanstd(stack,  axis=0)
        else:
            mean_curves[var][slope] = np.full(101, np.nan)
            std_curves[var][slope]  = np.full(101, np.nan)

# ============================================================
# HELPER: find peak within a band (max or min)
# ============================================================
def find_peak(mean_curve, band, mode='max'):
    mask = (stance_pct >= band[0]) & (stance_pct <= band[1])
    seg  = mean_curve[mask]
    sp   = stance_pct[mask]
    if len(seg) == 0 or np.all(np.isnan(seg)):
        return None, None
    idx = np.nanargmax(seg) if mode == 'max' else np.nanargmin(seg)
    return sp[idx], seg[idx]

# ============================================================
# COMPUTE TREND DIRECTION per variable per band
# ============================================================
def get_trend(var, band, mode='max'):
    """Returns (symbol, color) based on whether value increases or decreases
    from 0° to -18°."""
    val_flat  = find_peak(mean_curves[var].get(0,  np.full(101, np.nan)), band, mode)[1]
    val_steep = find_peak(mean_curves[var].get(-18, np.full(101, np.nan)), band, mode)[1]
    if val_flat is None or val_steep is None:
        return '', 'gray'
    if val_steep > val_flat:
        return '↑', '#c62828'   # red = increases with steeper slope
    else:
        return '↓', '#1565c0'   # blue = decreases with steeper slope

# ============================================================
# PLOT
# ============================================================
for var, ylabel in variables_to_plot.items():

    p1_mode = peak_logic.get(var, ('max', 'max'))[0]
    p2_mode = peak_logic.get(var, ('max', 'max'))[1]

    fig, ax = plt.subplots(figsize=(7.5, 5))

    # Phase shading
    ax.axvspan(*absorption_band, alpha=0.10, color='#2196F3', zorder=0)
    ax.axvspan(*propulsion_band, alpha=0.10, color='#FF7043', zorder=0)
    ax.text(np.mean(absorption_band), 0.97, 'Absorption',
            transform=ax.get_xaxis_transform(),
            ha='center', va='top', fontsize=8,
            color='#0d47a1', fontstyle='italic', fontweight='bold')
    ax.text(np.mean(propulsion_band), 0.97, 'Propulsion',
            transform=ax.get_xaxis_transform(),
            ha='center', va='top', fontsize=8,
            color='#bf360c', fontstyle='italic', fontweight='bold')

    # Lines + SD bands
    for slope in slopes_ordered:
        mean = mean_curves[var].get(slope)
        sd   = std_curves[var].get(slope)
        if mean is None or np.all(np.isnan(mean)):
            continue
        color = colors_slope[slope]
        ax.fill_between(stance_pct, mean - sd, mean + sd,
                        color=color, alpha=0.15, zorder=2)
        ax.plot(stance_pct, mean, color=color, linewidth=2.0,
                linestyle=linestyles[slope], zorder=3)

    # Circles at P1 / P2 per slope
    for slope in slopes_ordered:
        mean = mean_curves[var].get(slope)
        if mean is None or np.all(np.isnan(mean)):
            continue
        color = colors_slope[slope]

        px1, py1 = find_peak(mean, absorption_band, p1_mode)
        if px1 is not None:
            ax.plot(px1, py1, 'o', color=color, markersize=7,
                    markeredgecolor='white', markeredgewidth=1.2, zorder=6)

        px2, py2 = find_peak(mean, propulsion_band, p2_mode)
        if px2 is not None:
            ax.plot(px2, py2, 'o', color=color, markersize=7,
                    markeredgecolor='white', markeredgewidth=1.2, zorder=6)

    # Trend annotations (↑ / ↓) inside the bands
    ymin, ymax = ax.get_ylim()
    y_range    = ymax - ymin

    sym1, col1 = get_trend(var, absorption_band, p1_mode)
    sym2, col2 = get_trend(var, propulsion_band, p2_mode)

    # P1 label + trend
    ax.text(np.mean(absorption_band), ymin + y_range * 0.06,
            f'P1  {sym1}',
            ha='center', fontsize=9, fontweight='bold',
            color=col1, zorder=7)

    # P2 label + trend
    ax.text(np.mean(propulsion_band), ymin + y_range * 0.06,
            f'P2  {sym2}',
            ha='center', fontsize=9, fontweight='bold',
            color=col2, zorder=7)

    # Add a small note for GRF explaining the P2 decrease
    if var == 'GRF_vert':
        ax.text(0.98, 0.30,
                'P2 ↓ with steeper slope:\nreduced push-off demand\n(gravity assists propulsion)',
                transform=ax.transAxes, ha='right', va='top',
                fontsize=7.5, color='#1565c0', style='italic',
                bbox=dict(boxstyle='round,pad=0.4', facecolor='white',
                          edgecolor='#bbbbbb', alpha=0.85))

    # Slope legend
    slope_handles = [
        mlines.Line2D([], [], color=colors_slope[s], linestyle=linestyles[s],
                      linewidth=2.0, label=slope_labels[s])
        for s in slopes_ordered
    ]
    leg1 = ax.legend(handles=slope_handles, title='Slope', title_fontsize=8.5,
                     loc='upper right', fontsize=8.5,
                     framealpha=0.92, edgecolor='#cccccc')
    ax.add_artist(leg1)

    # Phase legend
    phase_handles = [
        mpatches.Patch(color='#2196F3', alpha=0.4,
                       label=f'Absorption ({absorption_band[0]}–{absorption_band[1]}%)'),
        mpatches.Patch(color='#FF7043', alpha=0.4,
                       label=f'Propulsion ({propulsion_band[0]}–{propulsion_band[1]}%)'),
    ]
    ax.legend(handles=phase_handles, loc='lower right',
              fontsize=7.5, framealpha=0.92, edgecolor='#cccccc')

    # Axes
    ax.set_xlabel('Gait Cycle (%)', fontsize=11, labelpad=5)
    ax.set_ylabel(ylabel, fontsize=10, labelpad=5)
    ax.set_title(ylabel.split('(')[0].strip(),
                 fontsize=12, fontweight='bold', pad=12)
    ax.set_xlim(0, 100)
    ax.set_xticks(range(0, 101, 25))
    ax.grid(True, linestyle=':', alpha=0.4, color='#aaaaaa')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.tick_params(labelsize=9)

    plt.tight_layout()
    plt.show()

# ============================================================
# PRINT PEAK VALUES PER BAND TO CONFIRM TRENDS
# ============================================================
print("\n" + "="*70)
print("PEAK VALUES PER PHASE BAND — confirming trend directions")
print("="*70)
print(f"\n{'Variable':<30} {'Band':<14} {'0°':>8} {'−6°':>8} {'−12°':>8} {'−18°':>8}  Trend")
print("-"*70)

for var, label in variables_to_plot.items():
    short = label.split('(')[0].strip()[:28]
    p1_mode = peak_logic.get(var, ('max','max'))[0]
    p2_mode = peak_logic.get(var, ('max','max'))[1]

    for band_name, band, mode in [('Absorption', absorption_band, p1_mode),
                                   ('Propulsion', propulsion_band, p2_mode)]:
        vals = []
        for slope in slopes_ordered:
            mean = mean_curves[var].get(slope, np.full(101, np.nan))
            _, pv = find_peak(mean, band, mode)
            vals.append(pv)

        trend = '↑' if (vals[-1] is not None and vals[0] is not None
                        and vals[-1] > vals[0]) else '↓'
        col   = '\033[31m' if trend == '↑' else '\033[34m'
        reset = '\033[0m'
        row   = [f"{v:.2f}" if v is not None else " nan" for v in vals]
        print(f"{short:<30} {band_name:<14} "
              f"{row[0]:>8} {row[1]:>8} {row[2]:>8} {row[3]:>8}  "
              f"{col}{trend}{reset}")

print()
print("↑ = increases with steeper downhill slope")
print("↓ = decreases with steeper downhill slope")
print("\nNote: GRF P2 ↓ is consistent with Alexander & Schwameder (2016):")
print("  'decreased vertical GRF at the second peak' during downhill walking")
```

```python
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd

# ============================================================
# FIXED HELPER FUNCTIONS
# ============================================================

def run_ols_with_subject(df_long, outcome_col):
    df_clean = df_long[['id', 'slope', outcome_col]].dropna().copy()
    # Force slope to plain Python int so formula keys are clean
    df_clean['slope'] = df_clean['slope'].astype(int)
    df_clean['id']    = df_clean['id'].astype(str)
    formula = f"{outcome_col} ~ C(slope, Treatment(0)) + C(id)"
    model   = smf.ols(formula, data=df_clean).fit()
    return model, df_clean


def extract_slope_results(model, slopes_vs):
    results = {}
    params  = model.params
    pvalues = model.pvalues
    tvalues = model.tvalues
    conf    = model.conf_int()
    df_res  = model.df_resid

    # Debug: show available keys if needed
    slope_keys = [k for k in params.index if 'slope' in k.lower()]

    for slope in slopes_vs:
        # Try both possible key formats
        key1 = f"C(slope, Treatment(0))[T.{int(slope)}]"
        key2 = f"C(slope, Treatment(0))[T.{slope}]"
        key  = key1 if key1 in params else (key2 if key2 in params else None)

        if key:
            coef       = params[key]
            p          = pvalues[key]
            t          = tvalues[key]
            ci_lo, ci_hi = conf.loc[key]
            dz         = abs(t) / np.sqrt(df_res + 1)

            if   dz >= 0.80: size = 'L'
            elif dz >= 0.50: size = 'M'
            elif dz >= 0.20: size = 'S'
            else:            size = 'trivial'

            results[slope] = {
                'coef':      coef,
                'p':         p,
                't':         t,
                'ci_lo':     ci_lo,
                'ci_hi':     ci_hi,
                'dz':        dz,
                'size':      size,
                'sig':       '*'  if p < alpha_bonferroni else 'ns',
                'direction': '↑'  if coef > 0 else '↓',
            }
        else:
            results[slope] = None
            print(f"    ⚠ Key not found for slope={slope}. Available: {slope_keys}")

    return results


# ============================================================
# STEP 10 — SIGNIFICANCE TABLE
# ============================================================
n_comparisons    = 3
alpha_bonferroni = 0.05 / n_comparisons
slopes_vs        = [-6, -12, -18]

for metric_label, suffix in [("MEAN OVER STANCE", "_mean"),
                              ("PEAK (MAXIMUM)",   "_peak")]:

    print("\n" + "="*105)
    print(f"OLS WITH SUBJECT AS FIXED COVARIATE — {metric_label}")
    print(f"Model: outcome ~ C(slope) + C(id)  |  Reference = 0°")
    print(f"Bonferroni α = 0.05 / {n_comparisons} = {alpha_bonferroni:.4f}")
    print("="*105)
    print(f"\n  {'Variable':<38} {'R²':>5}  "
          f"{'vs −6°':>24} {'vs −12°':>24} {'vs −18°':>24}")
    print("  " + "-"*115)

    for var, label in variables_to_plot.items():
        short = label.split('(')[0].strip()[:36]
        col   = f'{var}{suffix}'

        model, _ = run_ols_with_subject(df_long, col)
        res      = extract_slope_results(model, slopes_vs)
        r2       = model.rsquared

        cols_str = []
        for slope in slopes_vs:
            r = res.get(slope)
            if r is None:
                cols_str.append("              n/a")
            else:
                cols_str.append(
                    f"{r['direction']} p={r['p']:.4f}{r['sig']} "
                    f"dz={r['dz']:.2f}({r['size']})"
                )
        print(f"  {short:<38} {r2:>5.3f}  "
              f"{cols_str[0]:>24} {cols_str[1]:>24} {cols_str[2]:>24}")

    print(f"\n  * p < {alpha_bonferroni:.4f} (Bonferroni)")


# ============================================================
# STEP 11 — COEFFICIENTS TABLE with 95% CI
# ============================================================
print("\n\n" + "="*105)
print("COEFFICIENTS vs. LEVEL (0°)  —  Δ change in units  [95% CI]")
print("Controlling for participant ID as fixed covariate")
print("Forces in ×BW | Angles in ° | Moments in Nm")
print("="*105)

for metric_label, suffix in [("MEAN OVER STANCE", "_mean"),
                              ("PEAK (MAXIMUM)",   "_peak")]:

    print(f"\n  [{metric_label}]")
    print(f"  {'Variable':<38} "
          f"{'Δ −6° [95%CI]':>26} "
          f"{'Δ −12° [95%CI]':>26} "
          f"{'Δ −18° [95%CI]':>26}")
    print("  " + "-"*118)

    for var, label in variables_to_plot.items():
        short = label.split('(')[0].strip()[:36]
        col   = f'{var}{suffix}'

        model, _ = run_ols_with_subject(df_long, col)
        res      = extract_slope_results(model, slopes_vs)

        cols_str = []
        for slope in slopes_vs:
            r = res.get(slope)
            if r is None:
                cols_str.append("                   n/a")
            else:
                cols_str.append(
                    f"{r['coef']:+.3f} [{r['ci_lo']:+.3f}, {r['ci_hi']:+.3f}]{r['sig']}"
                )

        print(f"  {short:<38} "
              f"{cols_str[0]:>26} "
              f"{cols_str[1]:>26} "
              f"{cols_str[2]:>26}")

print(f"\n  * p < {alpha_bonferroni:.4f} (Bonferroni)")
print("  Δ = estimated change vs. level walking (0°), controlling for participant")
print("  Positive Δ = higher than level | Negative Δ = lower than level")
```

```python
# ============================================================
# BUILD df_peak1 and df_peak2 FROM df_merged
# Extracting scalar values at P1 (absorption) and P2 (propulsion)
# ============================================================

from scipy.interpolate import interp1d
import numpy as np
import pandas as pd

# Bands (% of stance)
absorption_band = (10, 30)   # P1
propulsion_band = (65, 85)   # P2

stance_pct = np.linspace(0, 100, 101)

# Variables with time series to extract peaks from
ts_vars = ['kneeFres', 'knee_PFc', 'GRF_vert', 'GRF_ap',
           'angleKnee', 'angleHip', 'angleAnkle',
           'momentKnee', 'momentHip', 'momentAnkle',
           'ankleFres', 'hipFres']

# Metadata columns to keep
meta_cols = ['id', 'file', 'side', 'slope', 'condition',
             'height', 'weight', 'age', 'speed',
             'step_length', 'stride_length', 'gc_dur',
             'stance', 'cadence']

def interp_stance(arr, stance_fraction):
    """Crop to stance phase and interpolate to 101 points."""
    arr = np.array(arr, dtype=float)
    n = len(arr)
    if n < 2:
        return np.full(101, np.nan)
    n_stance = max(2, round(stance_fraction * n))
    arr_s    = arr[:n_stance]
    x_orig   = np.linspace(0, 100, len(arr_s))
    f = interp1d(x_orig, arr_s, kind='linear',
                 bounds_error=False, fill_value=np.nan)
    return f(stance_pct)

def extract_peak_in_band(curve, band, mode='max'):
    """Return scalar peak value within a % stance band."""
    mask = (stance_pct >= band[0]) & (stance_pct <= band[1])
    seg  = curve[mask]
    if len(seg) == 0 or np.all(np.isnan(seg)):
        return np.nan
    return np.nanmax(seg) if mode == 'max' else np.nanmin(seg)

# Peak detection mode per variable per band
# GRF_ap at P1 is negative (braking) → use min
peak_modes = {
    'GRF_ap': ('min', 'max'),   # P1=braking(negative), P2=propulsion(positive)
}

rows_p1 = []
rows_p2 = []

for _, row in df_merged.iterrows():
    sf = row['stance']

    # Metadata
    meta = {col: row[col] for col in meta_cols if col in row.index}

    entry_p1 = meta.copy()
    entry_p2 = meta.copy()

    for var in ts_vars:
        if var not in row.index:
            entry_p1[var] = np.nan
            entry_p2[var] = np.nan
            continue

        val = row[var]
        if not isinstance(val, (list, np.ndarray)) or len(val) < 2:
            entry_p1[var] = np.nan
            entry_p2[var] = np.nan
            continue

        curve = interp_stance(val, sf)

        p1_mode, p2_mode = peak_modes.get(var, ('max', 'max'))
        entry_p1[var] = extract_peak_in_band(curve, absorption_band, p1_mode)
        entry_p2[var] = extract_peak_in_band(curve, propulsion_band, p2_mode)

    rows_p1.append(entry_p1)
    rows_p2.append(entry_p2)

df_peak1 = pd.DataFrame(rows_p1)
df_peak2 = pd.DataFrame(rows_p2)

# Fix types
for df_ in [df_peak1, df_peak2]:
    df_['id']    = df_['id'].astype(str).str.strip()
    df_['slope'] = df_['slope'].astype(int)

# Exclude participants 16 & 26
exclude  = ['16', '26']
df_peak1 = df_peak1[~df_peak1['id'].isin(exclude)].copy()
df_peak2 = df_peak2[~df_peak2['id'].isin(exclude)].copy()

# Normalize forces to ×BW
force_cols = ['kneeFres', 'knee_PFc', 'GRF_vert', 'GRF_ap',
              'ankleFres', 'hipFres']
for df_ in [df_peak1, df_peak2]:
    for col in force_cols:
        if col in df_.columns:
            df_[col] = df_[col] / (df_['weight'] * 9.81)

print("✓ df_peak1 and df_peak2 built from df_merged")
print(f"\n  PEAK 1 (Absorption band {absorption_band[0]}–{absorption_band[1]}% stance)")
print(f"    Rows         : {len(df_peak1)}")
print(f"    Participants : {df_peak1['id'].nunique()}")
print(f"    Slopes       : {sorted(df_peak1['slope'].unique())}")
print(f"    Conditions   : {df_peak1['condition'].unique()}")

print(f"\n  PEAK 2 (Propulsion band {propulsion_band[0]}–{propulsion_band[1]}% stance)")
print(f"    Rows         : {len(df_peak2)}")
print(f"    Participants : {df_peak2['id'].nunique()}")
print(f"    Slopes       : {sorted(df_peak2['slope'].unique())}")
print(f"    Conditions   : {df_peak2['condition'].unique()}")

print(f"\n  Sample means at 0° — PEAK 1:")
for var in ['kneeFres', 'knee_PFc', 'GRF_vert']:
    m = df_peak1[df_peak1['slope']==0][var].mean()
    print(f"    {var:<15}: {m:.3f} ×BW")

print(f"\n  Sample means at 0° — PEAK 2:")
for var in ['kneeFres', 'knee_PFc', 'GRF_vert']:
    m = df_peak2[df_peak2['slope']==0][var].mean()
    print(f"    {var:<15}: {m:.3f} ×BW")
```

```python
# ============================================================
# OLS WITH SUBJECT AS FIXED COVARIATE — PEAK 1 vs PEAK 2
# ============================================================

import statsmodels.formula.api as smf

variables_scalar = {
    'kneeFres':   'Tibiofemoral Joint Compression Force (×BW)',
    'knee_PFc':   'Patellofemoral Joint Compression Force (×BW)',
    'GRF_vert':   'Vertical Ground Reaction Force (×BW)',
    'angleKnee':  'Knee Flexion Angle (°)',
    'momentKnee': 'Knee Extension Moment (Nm)',
}

slopes_vs        = [-6, -12, -18]
n_comparisons    = 3
alpha_bonferroni = 0.05 / n_comparisons  # 0.0167


def run_ols_scalar(df, outcome_col):
    df_clean = df[['id', 'slope', outcome_col]].dropna().copy()
    df_clean['slope'] = df_clean['slope'].astype(int)
    df_clean['id']    = df_clean['id'].astype(str)
    formula = f"{outcome_col} ~ C(slope, Treatment(0)) + C(id)"
    return smf.ols(formula, data=df_clean).fit()


def extract_results(model, slopes_vs):
    params  = model.params
    pvalues = model.pvalues
    tvalues = model.tvalues
    conf    = model.conf_int()
    df_res  = model.df_resid
    results = {}

    for slope in slopes_vs:
        key = f"C(slope, Treatment(0))[T.{int(slope)}]"
        if key in params:
            coef         = params[key]
            p            = pvalues[key]
            t            = tvalues[key]
            ci_lo, ci_hi = conf.loc[key]
            dz           = abs(t) / np.sqrt(df_res + 1)

            if   dz >= 0.80: size = 'L'
            elif dz >= 0.50: size = 'M'
            elif dz >= 0.20: size = 'S'
            else:            size = 'trivial'

            results[slope] = {
                'coef': coef, 'p': p,
                'ci_lo': ci_lo, 'ci_hi': ci_hi,
                'dz': dz, 'size': size,
                'sig': '*' if p < alpha_bonferroni else 'ns',
                'direction': '↑' if coef > 0 else '↓',
            }
        else:
            results[slope] = None
    return results


# ============================================================
# PRINT RESULTS
# ============================================================

for peak_label, df_ in [
        ("PEAK 1 — Absorption phase (10–30% stance)", df_peak1),
        ("PEAK 2 — Propulsion phase (65–85% stance)", df_peak2)]:

    print("\n" + "="*108)
    print(f"OLS — {peak_label}")
    print(f"Model: value ~ C(slope) + C(id)  |  Reference = 0°  |  Excl: 16 & 26")
    print(f"Bonferroni α = {alpha_bonferroni:.4f}  |  "
          f"Effect: L>0.80 | M=0.50–0.79 | S=0.20–0.49 | trivial<0.20")
    print("="*108)

    # --- Significance table ---
    print(f"\n  {'Variable':<38} {'R²':>5}  "
          f"{'vs −6°':>24} {'vs −12°':>24} {'vs −18°':>24}")
    print("  " + "-"*118)

    for var, label in variables_scalar.items():
        if var not in df_.columns:
            continue
        short = label.split('(')[0].strip()[:36]
        model = run_ols_scalar(df_, var)
        res   = extract_results(model, slopes_vs)
        r2    = model.rsquared

        cols_str = []
        for slope in slopes_vs:
            r = res.get(slope)
            if r is None:
                cols_str.append("                   n/a")
            else:
                cols_str.append(
                    f"{r['direction']} p={r['p']:.4f}{r['sig']} "
                    f"dz={r['dz']:.2f}({r['size']})"
                )
        print(f"  {short:<38} {r2:>5.3f}  "
              f"{cols_str[0]:>24} {cols_str[1]:>24} {cols_str[2]:>24}")

    # --- Coefficients table ---
    print(f"\n  Δ vs. level (0°)  [95% CI]")
    print(f"  {'Variable':<38} "
          f"{'Δ −6° [95%CI]':>28} "
          f"{'Δ −12° [95%CI]':>28} "
          f"{'Δ −18° [95%CI]':>28}")
    print("  " + "-"*122)

    for var, label in variables_scalar.items():
        if var not in df_.columns:
            continue
        short = label.split('(')[0].strip()[:36]
        model = run_ols_scalar(df_, var)
        res   = extract_results(model, slopes_vs)

        cols_str = []
        for slope in slopes_vs:
            r = res.get(slope)
            if r is None:
                cols_str.append("                         n/a")
            else:
                cols_str.append(
                    f"{r['coef']:+.3f} "
                    f"[{r['ci_lo']:+.3f}, {r['ci_hi']:+.3f}]{r['sig']}"
                )
        print(f"  {short:<38} "
              f"{cols_str[0]:>28} "
              f"{cols_str[1]:>28} "
              f"{cols_str[2]:>28}")

    # --- Descriptive means per slope ---
    print(f"\n  Mean ± SD per slope (×BW or units)")
    print(f"  {'Variable':<38} {'0°':>10} {'−6°':>10} {'−12°':>10} {'−18°':>10}")
    print("  " + "-"*80)

    for var, label in variables_scalar.items():
        if var not in df_.columns:
            continue
        short   = label.split('(')[0].strip()[:36]
        row_str = []
        for slope in [0, -6, -12, -18]:
            vals = df_[df_['slope'] == slope][var].dropna()
            m = vals.mean() if len(vals) > 0 else np.nan
            s = vals.std()  if len(vals) > 0 else np.nan
            row_str.append(f"{m:.2f}±{s:.2f}")
        print(f"  {short:<38} "
              f"{row_str[0]:>10} {row_str[1]:>10} "
              f"{row_str[2]:>10} {row_str[3]:>10}")

    print(f"\n  * p < {alpha_bonferroni:.4f} (Bonferroni)")
    print("  Δ = estimated change vs. 0°, controlling for participant ID")

# ============================================================
# ALEXANDER 2016 REFERENCE
# ============================================================
print("\n\n" + "="*80)
print("REFERENCE — Alexander & Schwameder (2016)")
print("="*80)
print(f"\n  {'Variable':<38} {'0°':>10} {'−6°':>10} {'−12°':>10} {'−18°':>10}")
print("  " + "-"*80)
print(f"  {'[Table 1] Tibiofemoral mean':<38} "
      f"{'2.38':>10} {'2.32':>10} {'2.81':>10} {'3.04':>10}")
print(f"  {'[Table 1] Patellofemoral mean':<38} "
      f"{'0.23':>10} {'0.54':>10} {'1.36':>10} {'2.03':>10}")
print(f"  {'[Table 2] Tibiofemoral peak':<38} "
      f"{'4.09':>10} {'3.77':>10} {'4.88':>10} {'5.50':>10}")
print(f"  {'[Table 2] Patellofemoral peak':<38} "
      f"{'0.86':>10} {'1.46':>10} {'2.78':>10} {'3.96':>10}")
print("\n  Note: Alexander reports stance-phase mean (Table 1) and maximum (Table 2).")
print("  Your P1 values correspond best to Table 2 peaks.")
print("  Your P2 patellofemoral (0.215 ×BW at 0°) matches Alexander Table 1 (0.23 ×BW).")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# ============================================================
# Z-SCORE OUTLIER DETECTION — Peak 1
# ============================================================

variables_check = {
    'kneeFres':   'Tibiofemoral (×BW)',
    'knee_PFc':   'Patellofemoral (×BW)',
    'GRF_vert':   'GRF Vertical (×BW)',
    'angleKnee':  'Knee Angle (°)',
    'momentKnee': 'Knee Moment (Nm)',
    'speed':      'Speed (m/s)',
}

# Aggregate to 1 value per participant (mean across slopes and trials)
df_agg = (df_peak1.groupby('id')[list(variables_check.keys())]
          .mean()
          .reset_index())

# Compute z-scores
df_z = df_agg.copy()
for var in variables_check:
    df_z[f'z_{var}'] = stats.zscore(df_agg[var], nan_policy='omit')

# ============================================================
# PRINT Z-SCORE TABLE
# ============================================================
z_threshold = 2.0

print("="*95)
print(f"Z-SCORE TABLE — Peak 1  (|z| > {z_threshold} flagged with *)")
print("="*95)

# Header
header = f"  {'ID':<6}"
for var, label in variables_check.items():
    header += f" {label:>18}"
print(header)
print("  " + "-"*90)

flag_counts = {}
for _, row in df_z.iterrows():
    pid    = row['id']
    line   = f"  {pid:<6}"
    flags  = 0
    for var, label in variables_check.items():
        z = row[f'z_{var}']
        if np.isnan(z):
            line  += f" {'nan':>18}"
        elif abs(z) > z_threshold:
            line  += f" {z:>17.2f}*"
            flags += 1
        else:
            line  += f" {z:>18.2f}"
    flag_counts[pid] = flags
    marker = '  ⚠' if flags >= 2 else ''
    print(line + marker)

print("  " + "-"*90)
print(f"  * |z| > {z_threshold}  |  ⚠ = flagged on ≥2 variables")

# Summary
flagged = [pid for pid, n in flag_counts.items() if n >= 2]
print(f"\n  Participants flagged on ≥2 variables: "
      f"{flagged if flagged else 'None'}")

# ============================================================
# HEATMAP
# ============================================================
z_cols = [f'z_{var}' for var in variables_check]
z_matrix = df_z.set_index('id')[z_cols].values
ids      = df_z['id'].values
labels   = list(variables_check.values())

fig, ax = plt.subplots(figsize=(10, max(5, len(ids) * 0.4 + 1)))

im = ax.imshow(z_matrix, cmap='RdBu_r', vmin=-3, vmax=3,
               aspect='auto')

# Axes
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=30, ha='right', fontsize=9)
ax.set_yticks(range(len(ids)))
ax.set_yticklabels(ids, fontsize=9)

# Cell values
for i in range(len(ids)):
    for j in range(len(labels)):
        val  = z_matrix[i, j]
        mark = '*' if abs(val) > z_threshold else ''
        txt  = f"{val:.2f}{mark}" if not np.isnan(val) else 'nan'
        color = 'white' if abs(val) > 1.8 else 'black'
        ax.text(j, i, txt, ha='center', va='center',
                fontsize=8, color=color, fontweight='bold' if mark else 'normal')

plt.colorbar(im, ax=ax, label='Z-score', shrink=0.8)
ax.set_title('Z-score Heatmap — Peak 1\n'
             f'Red = high | Blue = low | * = |z| > {z_threshold}',
             fontsize=11, fontweight='bold', pad=12)

plt.tight_layout()
fig.savefig("figures/zscore_heatmap_peak1.png",
            dpi=300, bbox_inches='tight', facecolor='white')
print("\n✓ Saved: figures/zscore_heatmap_peak1.png")
plt.show()
```

```python
# ============================================================
# SENSITIVITY ANALYSIS — with vs without participant 20
# ============================================================

targets_check = ['kneeFres', 'knee_PFc', 'GRF_vert', 'angleKnee', 'momentKnee']
slopes_vs     = [-6, -12, -18]

def run_ols_scalar(df, outcome_col):
    df_clean = df[['id', 'slope', outcome_col]].dropna().copy()
    df_clean['slope'] = df_clean['slope'].astype(int)
    df_clean['id']    = df_clean['id'].astype(str)
    formula = f"{outcome_col} ~ C(slope, Treatment(0)) + C(id)"
    return smf.ols(formula, data=df_clean).fit()

def extract_coef(model, slopes_vs):
    results = {}
    for slope in slopes_vs:
        key = f"C(slope, Treatment(0))[T.{int(slope)}]"
        if key in model.params:
            results[slope] = {
                'coef': model.params[key],
                'p':    model.pvalues[key],
                'r2':   model.rsquared,
            }
    return results

df_full    = df_peak1.copy()
df_no20    = df_peak1[df_peak1['id'] != '20'].copy()

print("="*90)
print("SENSITIVITY ANALYSIS — Peak 1:  Full dataset vs. Excluding Participant 20")
print("="*90)

for var in targets_check:
    label = variables_check.get(var, var)

    m_full = run_ols_scalar(df_full, var)
    m_no20 = run_ols_scalar(df_no20, var)

    res_full = extract_coef(m_full, slopes_vs)
    res_no20 = extract_coef(m_no20, slopes_vs)

    print(f"\n  {label}")
    print(f"  {'':>6} {'R² full':>9} {'R² -P20':>9} {'ΔR²':>7}")
    print(f"  {'':>6} {m_full.rsquared:>9.3f} {m_no20.rsquared:>9.3f} "
          f"{m_no20.rsquared - m_full.rsquared:>+7.3f}")

    print(f"\n  {'Slope':>8} {'Δ full':>10} {'p full':>8} "
          f"{'Δ -P20':>10} {'p -P20':>8} {'Δ coef':>9}  {'Stable?'}")
    print("  " + "-"*72)

    for slope in slopes_vs:
        rf = res_full.get(slope, {})
        rn = res_no20.get(slope, {})
        if not rf or not rn:
            continue

        delta_coef = rn['coef'] - rf['coef']
        # Stable if: same direction, similar magnitude (< 20% change), same significance
        same_dir   = np.sign(rf['coef']) == np.sign(rn['coef'])
        pct_change = abs(delta_coef / rf['coef']) * 100 if rf['coef'] != 0 else np.nan
        both_sig   = (rf['p'] < 0.0167) == (rn['p'] < 0.0167)
        stable     = '✓ stable' if (same_dir and pct_change < 20 and both_sig) else '⚠ changes'

        print(f"  {slope:>8}° {rf['coef']:>+10.3f} {rf['p']:>8.4f} "
              f"{rn['coef']:>+10.3f} {rn['p']:>8.4f} "
              f"{delta_coef:>+9.3f}  {stable}  ({pct_change:.1f}% change)")

# Mean comparison
print("\n\n" + "="*90)
print("MEAN ± SD COMPARISON — 0° and −18°")
print("="*90)
print(f"\n  {'Variable':<30} {'Full 0°':>12} {'−P20 0°':>12} "
      f"{'Full −18°':>12} {'−P20 −18°':>12}")
print("  " + "-"*72)

for var in targets_check:
    label = variables_check.get(var, var)[:28]
    for slope, s_label in [(0, '0°'), (-18, '−18°')]:
        pass  # build below

    v_full_0   = df_full[df_full['slope']==0][var].dropna()
    v_no20_0   = df_no20[df_no20['slope']==0][var].dropna()
    v_full_18  = df_full[df_full['slope']==-18][var].dropna()
    v_no20_18  = df_no20[df_no20['slope']==-18][var].dropna()

    print(f"  {label:<30} "
          f"{v_full_0.mean():>5.2f}±{v_full_0.std():.2f} "
          f"{v_no20_0.mean():>5.2f}±{v_no20_0.std():.2f} "
          f"{v_full_18.mean():>5.2f}±{v_full_18.std():.2f} "
          f"{v_no20_18.mean():>5.2f}±{v_no20_18.std():.2f}")

print("\n  ✓ stable  = same direction, <20% coefficient change, same significance")
print("  ⚠ changes = direction flip, >20% change, or significance changes")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import statsmodels.formula.api as smf

# ============================================================
# CONFIGURATION
# ============================================================
df_p1 = df_peak1[df_peak1['id'] != '20'].copy()
df_p1['id']    = df_p1['id'].astype(str).str.strip()
df_p1['slope'] = df_p1['slope'].astype(int)

model_config = {
    'kneeFres': {
        'features': ['angleKnee', 'angleHip',
                     'height', 'weight', 'speed', 'slope'],
        'label':    'Tibiofemoral Joint Compression Force (×BW)',
        'scenario': 'No Cadence & No Ankle Angle',
    },
    'knee_PFc': {
        'features': ['angleKnee', 'angleHip', 'angleAnkle',
                     'height', 'weight', 'speed', 'slope'],
        'label':    'Patellofemoral Joint Compression Force (×BW)',
        'scenario': 'No Cadence',
    },
}

slopes_ordered = [0, -6, -12, -18]
colors_slope   = {0: '#a8d5e2', -6: '#5b9fbf', -12: '#2a6a96', -18: '#0d2d55'}

# ============================================================
# FIT + PLOT
# ============================================================
results_summary = []

fig, axes = plt.subplots(1, 2, figsize=(16, 7))
fig.suptitle('Linear Regression — Peak 1 (Absorption Phase)\n'
             'Subject ID as Fixed Covariate | Participant 20 Excluded',
             fontsize=13, fontweight='bold')

for ax, (target, cfg) in zip(axes, model_config.items()):

    features = cfg['features']
    label    = cfg['label']

    # --- Clean data: avoid duplicate columns ---
    # slope is already in features, don't add it again
    all_cols = list(dict.fromkeys(['id', 'slope', target] + features))
    df_c     = df_p1[all_cols].dropna().copy().reset_index(drop=True)
    df_c['slope'] = df_c['slope'].astype(int)

    # --- Build X matrix ---
    subj_dummies = pd.get_dummies(df_c['id'], prefix='Subj', drop_first=True)
    X = pd.concat([df_c[features].reset_index(drop=True),
                   subj_dummies.reset_index(drop=True)], axis=1)
    y = df_c[target].values.astype(float)   # plain 1D numpy array

    # --- Fit ---
    model     = LinearRegression()
    model.fit(X.values, y)
    y_pred    = model.predict(X.values)     # plain 1D numpy array
    slope_arr = df_c['slope'].values.astype(int)  # plain 1D numpy array

    # quick sanity check
    assert y.ndim == 1, f"y is not 1D: {y.shape}"
    assert y_pred.ndim == 1, f"y_pred is not 1D: {y_pred.shape}"
    assert slope_arr.ndim == 1, f"slope_arr is not 1D: {slope_arr.shape}"

    # --- Stats ---
    r2       = r2_score(y, y_pred)
    rmse     = np.sqrt(mean_squared_error(y, y_pred))
    rmse_pct = (rmse / y.mean()) * 100
    n        = len(y)

    # --- Equation string ---
    eq_lines = [f"y = {model.intercept_:.2f}"]
    for name, coef in zip(X.columns, model.coef_):
        if 'Subj' not in name:
            clean = name.replace('angle', 'Ang')
            eq_lines.append(f"{coef:+.2f} × {clean}")
    eq_lines.append("[+ Subject Effects]")
    eq_str = "\n".join(eq_lines)

    results_summary.append({
        'Target':   target,
        'Scenario': cfg['scenario'],
        'n':        n,
        'R²':       round(r2, 3),
        'RMSE (%)': round(rmse_pct, 2),
        'Features': ", ".join(features),
    })

    # --- Scatter ---
    for slope in slopes_ordered:
        idx = np.where(slope_arr == slope)[0]  # explicit integer indices
        if len(idx) == 0:
            continue
        ax.scatter(y[idx], y_pred[idx],
                   color=colors_slope[slope], alpha=0.7, s=55,
                   label=f'{slope}°', edgecolors='white',
                   linewidths=0.4, zorder=3)

    # Perfect fit line
    vmin = min(y.min(), y_pred.min())
    vmax = max(y.max(), y_pred.max())
    pad  = (vmax - vmin) * 0.05
    ax.plot([vmin - pad, vmax + pad],
            [vmin - pad, vmax + pad],
            color='red', linestyle='--', linewidth=1.8,
            label='Perfect Fit', zorder=2)

    # Stats box
    stats_text = (f"n = {n}\n"
                  f"$R^2$ = {r2:.3f}\n"
                  f"RMSE = {rmse_pct:.1f}%\n\n"
                  f"{eq_str}")
    ax.text(0.03, 0.97, stats_text,
            transform=ax.transAxes, fontsize=7.5,
            verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='white',
                      alpha=0.9, edgecolor='#cccccc'))

    # Formatting
    ax.set_xlabel(f"Actual {label}", fontsize=10)
    ax.set_ylabel(f"Predicted {label}", fontsize=10)
    ax.set_title(f"Target: {label.split('(')[0].strip()}",
                 fontsize=11, fontweight='bold')
    ax.set_xlim(vmin - pad, vmax + pad)
    ax.set_ylim(vmin - pad, vmax + pad)
    ax.grid(True, linestyle=':', alpha=0.4)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.tick_params(labelsize=9)

    handles, lbls = ax.get_legend_handles_labels()
    ax.legend(handles, lbls, title='Slope (°)', fontsize=8,
              title_fontsize=8.5, loc='lower right',
              framealpha=0.9, edgecolor='#cccccc')

plt.tight_layout()
fig.savefig("figures/regression_peak1.png",
            dpi=300, bbox_inches='tight', facecolor='white')
print("✓ Saved: figures/regression_peak1.png")
plt.show()

# ============================================================
# SUMMARY TABLE
# ============================================================
print("\n" + "="*85)
print("REGRESSION SUMMARY — Peak 1  (Participant 20 excluded)")
print("="*85)
print(pd.DataFrame(results_summary).to_string(index=False))

# ============================================================
# COEFFICIENTS WITH p-VALUES (statsmodels)
# ============================================================
print("\n" + "="*85)
print("COEFFICIENTS WITH p-VALUES (statsmodels OLS)")
print("="*85)

for target, cfg in model_config.items():
    features = cfg['features']
    label    = cfg['label'].split('(')[0].strip()

    all_cols = list(dict.fromkeys(['id', 'slope', target] + features))
    df_c     = df_p1[all_cols].dropna().copy().reset_index(drop=True)
    df_c['slope'] = df_c['slope'].astype(int)
    df_c['id']    = df_c['id'].astype(str)

    formula  = f"{target} ~ " + " + ".join(features) + " + C(id)"
    sm_model = smf.ols(formula, data=df_c).fit()

    print(f"\n  [{label}]")
    print(f"  R² = {sm_model.rsquared:.3f}  |  "
          f"Adj. R² = {sm_model.rsquared_adj:.3f}  |  "
          f"n = {int(sm_model.nobs)}")
    print(f"\n  {'Variable':<20} {'Coef':>10} {'SE':>8} "
          f"{'t':>8} {'p':>10}  Sig")
    print("  " + "-"*65)

    for name in sm_model.params.index:
        if 'C(id)' in name:
            continue
        coef = sm_model.params[name]
        se   = sm_model.bse[name]
        t    = sm_model.tvalues[name]
        p    = sm_model.pvalues[name]
        sig  = ('***' if p < 0.001 else
                '**'  if p < 0.01  else
                '*'   if p < 0.05  else 'ns')
        clean = name.replace('angle', 'Ang')
        print(f"  {clean:<20} {coef:>10.3f} {se:>8.3f} "
              f"{t:>8.3f} {p:>10.4f}  {sig}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import statsmodels.formula.api as smf
from scipy.interpolate import interp1d
from scipy import stats

# ============================================================
# STEP 1 — REBUILD df_peak2 WITH ADAPTIVE BANDS PER SLOPE
# ============================================================

# Adaptive propulsion bands per slope
# Level: P2 occurs earlier (~45–65%)
# Steeper downhill: P2 occurs later (~65–85%)
propulsion_bands = {
     0: (45, 65),
    -6: (50, 70),
   -12: (60, 80),
   -18: (65, 85),
}

stance_pct = np.linspace(0, 100, 101)

def interp_stance(arr, stance_fraction):
    arr = np.array(arr, dtype=float)
    n = len(arr)
    if n < 2:
        return np.full(101, np.nan)
    n_stance = max(2, round(stance_fraction * n))
    arr_s    = arr[:n_stance]
    x_orig   = np.linspace(0, 100, len(arr_s))
    f = interp1d(x_orig, arr_s, kind='linear',
                 bounds_error=False, fill_value=np.nan)
    return f(stance_pct)

def extract_peak_adaptive(arr, stance_fraction, slope):
    """Extract peak value using slope-specific propulsion band."""
    curve = interp_stance(arr, stance_fraction)
    band  = propulsion_bands.get(int(slope), (50, 85))
    mask  = (stance_pct >= band[0]) & (stance_pct <= band[1])
    seg   = curve[mask]
    if len(seg) == 0 or np.all(np.isnan(seg)):
        return np.nan
    return np.nanmax(seg)

# Variables to extract
ts_vars  = ['kneeFres', 'knee_PFc', 'GRF_vert', 'GRF_ap',
            'angleKnee', 'angleHip', 'angleAnkle',
            'momentKnee', 'momentHip', 'momentAnkle',
            'ankleFres', 'hipFres']
meta_cols = ['id', 'file', 'side', 'slope', 'condition',
             'height', 'weight', 'age', 'speed',
             'step_length', 'stride_length', 'gc_dur',
             'stance', 'cadence']

rows_p2 = []
for _, row in df_merged.iterrows():
    sf    = row['stance']
    slope = int(row['slope'])
    meta  = {col: row[col] for col in meta_cols if col in row.index}
    entry = meta.copy()
    for var in ts_vars:
        if var not in row.index:
            entry[var] = np.nan
            continue
        val = row[var]
        if not isinstance(val, (list, np.ndarray)) or len(val) < 2:
            entry[var] = np.nan
            continue
        # GRF_ap at P2 is positive (propulsion) → max
        entry[var] = extract_peak_adaptive(val, sf, slope)
    rows_p2.append(entry)

df_peak2_adaptive = pd.DataFrame(rows_p2)
df_peak2_adaptive['id']    = df_peak2_adaptive['id'].astype(str).str.strip()
df_peak2_adaptive['slope'] = df_peak2_adaptive['slope'].astype(int)

# Exclude participants 16, 26, 20
exclude = ['16', '26', '20']
df_peak2_adaptive = df_peak2_adaptive[
    ~df_peak2_adaptive['id'].isin(exclude)].copy()

# Normalize forces to ×BW
force_cols = ['kneeFres', 'knee_PFc', 'GRF_vert', 'GRF_ap',
              'ankleFres', 'hipFres']
for col in force_cols:
    if col in df_peak2_adaptive.columns:
        df_peak2_adaptive[col] = (df_peak2_adaptive[col] /
                                  (df_peak2_adaptive['weight'] * 9.81))

print("✓ df_peak2_adaptive built with adaptive propulsion bands")
print(f"  Rows         : {len(df_peak2_adaptive)}")
print(f"  Participants : {df_peak2_adaptive['id'].nunique()}")
print(f"  Slopes       : {sorted(df_peak2_adaptive['slope'].unique())}")
print(f"\n  Bands used per slope:")
for s, b in propulsion_bands.items():
    print(f"    {s:>4}° → {b[0]}–{b[1]}% stance")

print(f"\n  Sanity check — Mean ± SD at 0°:")
for var in ['kneeFres', 'knee_PFc', 'GRF_vert']:
    vals = df_peak2_adaptive[df_peak2_adaptive['slope']==0][var].dropna()
    print(f"    {var:<15}: {vals.mean():.3f} ± {vals.std():.3f} ×BW")

# ============================================================
# STEP 2 — Z-SCORE CHECK
# ============================================================
variables_check = {
    'kneeFres':   'Tibiofemoral (×BW)',
    'knee_PFc':   'Patellofemoral (×BW)',
    'GRF_vert':   'GRF Vertical (×BW)',
    'angleKnee':  'Knee Angle (°)',
    'momentKnee': 'Knee Moment (Nm)',
}

df_agg_p2 = (df_peak2_adaptive.groupby('id')[list(variables_check.keys())]
             .mean().reset_index())
df_z2 = df_agg_p2.copy()
for var in variables_check:
    df_z2[f'z_{var}'] = stats.zscore(df_agg_p2[var], nan_policy='omit')

z_threshold = 2.0
print("\n" + "="*85)
print(f"Z-SCORE TABLE — Peak 2 Adaptive  (|z| > {z_threshold} flagged *)")
print("="*85)
header = f"  {'ID':<6}"
for label in variables_check.values():
    header += f" {label:>20}"
print(header)
print("  " + "-"*85)

flag_counts = {}
for _, row in df_z2.iterrows():
    pid = row['id']
    line  = f"  {pid:<6}"
    flags = 0
    for var in variables_check:
        z = row[f'z_{var}']
        if np.isnan(z):
            line += f" {'nan':>20}"
        elif abs(z) > z_threshold:
            line += f" {z:>19.2f}*"
            flags += 1
        else:
            line += f" {z:>20.2f}"
    flag_counts[pid] = flags
    marker = '  ⚠' if flags >= 2 else ''
    print(line + marker)

print("  " + "-"*85)
flagged_p2 = [pid for pid, n in flag_counts.items() if n >= 2]
print(f"\n  Participants flagged ≥2 variables: "
      f"{flagged_p2 if flagged_p2 else 'None'}")

# ============================================================
# STEP 3 — REGRESSION
# ============================================================
exclude_final = list(set(exclude + flagged_p2))
df_p2 = df_peak2_adaptive[
    ~df_peak2_adaptive['id'].isin(exclude_final)].copy()
df_p2['id']    = df_p2['id'].astype(str).str.strip()
df_p2['slope'] = df_p2['slope'].astype(int)

print(f"\n  Final exclusions: {exclude_final}")
print(f"  Participants in model: {df_p2['id'].nunique()}")

model_config_p2 = {
    'kneeFres': {
        'features': ['angleKnee', 'angleHip',
                     'height', 'weight', 'speed', 'slope'],
        'label':    'Tibiofemoral Joint Compression Force (×BW)',
    },
    'knee_PFc': {
        'features': ['angleKnee', 'angleHip', 'angleAnkle',
                     'height', 'weight', 'speed', 'slope'],
        'label':    'Patellofemoral Joint Compression Force (×BW)',
    },
}

slopes_ordered = [0, -6, -12, -18]
colors_slope   = {0: '#a8d5e2', -6: '#5b9fbf', -12: '#2a6a96', -18: '#0d2d55'}

results_summary_p2 = []
fig, axes = plt.subplots(1, 2, figsize=(16, 7))
fig.suptitle('Linear Regression — Peak 2 (Propulsion Phase, Adaptive Bands)\n'
             f'Subject ID as Fixed Covariate | Excluded: {exclude_final}',
             fontsize=13, fontweight='bold')

for ax, (target, cfg) in zip(axes, model_config_p2.items()):
    features = cfg['features']
    label    = cfg['label']

    all_cols = list(dict.fromkeys(['id', 'slope', target] + features))
    df_c     = df_p2[all_cols].dropna().copy().reset_index(drop=True)
    df_c['slope'] = df_c['slope'].astype(int)

    subj_dummies = pd.get_dummies(df_c['id'], prefix='Subj', drop_first=True)
    X = pd.concat([df_c[features].reset_index(drop=True),
                   subj_dummies.reset_index(drop=True)], axis=1)
    y         = df_c[target].values.astype(float)
    slope_arr = df_c['slope'].values.astype(int)

    model  = LinearRegression()
    model.fit(X.values, y)
    y_pred = model.predict(X.values)

    r2       = r2_score(y, y_pred)
    rmse     = np.sqrt(mean_squared_error(y, y_pred))
    rmse_pct = (rmse / y.mean()) * 100
    n        = len(y)

    eq_lines = [f"y = {model.intercept_:.2f}"]
    for name, coef in zip(X.columns, model.coef_):
        if 'Subj' not in name:
            clean = name.replace('angle', 'Ang')
            eq_lines.append(f"{coef:+.2f} × {clean}")
    eq_lines.append("[+ Subject Effects]")
    eq_str = "\n".join(eq_lines)

    results_summary_p2.append({
        'Target':   target,
        'n':        n,
        'R²':       round(r2, 3),
        'RMSE (%)': round(rmse_pct, 2),
        'Features': ", ".join(features),
    })

    for slope in slopes_ordered:
        idx = np.where(slope_arr == slope)[0]
        if len(idx) == 0:
            continue
        ec = '#666666' if slope == 0 else 'white'
        ax.scatter(y[idx], y_pred[idx],
                   color=colors_slope[slope], alpha=0.7, s=55,
                   label=f'{slope}°', edgecolors=ec,
                   linewidths=0.6, zorder=3)

    vmin = min(y.min(), y_pred.min())
    vmax = max(y.max(), y_pred.max())
    pad  = (vmax - vmin) * 0.05
    ax.plot([vmin - pad, vmax + pad],
            [vmin - pad, vmax + pad],
            color='red', linestyle='--', linewidth=1.8,
            label='Perfect Fit', zorder=2)

    stats_text = (f"n = {n}\n$R^2$ = {r2:.3f}\nRMSE = {rmse_pct:.1f}%"
                  f"\n\n{eq_str}")
    ax.text(0.03, 0.97, stats_text, transform=ax.transAxes,
            fontsize=7.5, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='white',
                      alpha=0.9, edgecolor='#cccccc'))

    ax.set_xlabel(f"Actual {label}", fontsize=10)
    ax.set_ylabel(f"Predicted {label}", fontsize=10)
    ax.set_title(f"Target: {label.split('(')[0].strip()}",
                 fontsize=11, fontweight='bold')
    ax.set_xlim(vmin - pad, vmax + pad)
    ax.set_ylim(vmin - pad, vmax + pad)
    ax.grid(True, linestyle=':', alpha=0.4)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.tick_params(labelsize=9)
    handles, lbls = ax.get_legend_handles_labels()
    ax.legend(handles, lbls, title='Slope (°)', fontsize=8,
              title_fontsize=8.5, loc='lower right',
              framealpha=0.9, edgecolor='#cccccc')

plt.tight_layout()
fig.savefig("figures/regression_peak2_adaptive.png",
            dpi=300, bbox_inches='tight', facecolor='white')
print("\n✓ Saved: figures/regression_peak2_adaptive.png")
plt.show()

# ============================================================
# STEP 4 — SUMMARY + COEFFICIENTS
# ============================================================
print("\n" + "="*85)
print("REGRESSION SUMMARY — Peak 2 Adaptive")
print("="*85)
print(pd.DataFrame(results_summary_p2).to_string(index=False))

print("\n" + "="*85)
print("COEFFICIENTS WITH p-VALUES — Peak 2 Adaptive")
print("="*85)

for target, cfg in model_config_p2.items():
    features = cfg['features']
    label    = cfg['label'].split('(')[0].strip()
    all_cols = list(dict.fromkeys(['id', 'slope', target] + features))
    df_c     = df_p2[all_cols].dropna().copy().reset_index(drop=True)
    df_c['slope'] = df_c['slope'].astype(int)
    df_c['id']    = df_c['id'].astype(str)
    formula  = f"{target} ~ " + " + ".join(features) + " + C(id)"
    sm_model = smf.ols(formula, data=df_c).fit()

    print(f"\n  [{label}]")
    print(f"  R² = {sm_model.rsquared:.3f} | "
          f"Adj. R² = {sm_model.rsquared_adj:.3f} | "
          f"n = {int(sm_model.nobs)}")
    print(f"\n  {'Variable':<20} {'Coef':>10} {'SE':>8} "
          f"{'t':>8} {'p':>10}  Sig")
    print("  " + "-"*65)
    for name in sm_model.params.index:
        if 'C(id)' in name:
            continue
        coef = sm_model.params[name]
        se   = sm_model.bse[name]
        t    = sm_model.tvalues[name]
        p    = sm_model.pvalues[name]
        sig  = ('***' if p < 0.001 else '**' if p < 0.01
                else '*' if p < 0.05 else 'ns')
        clean = name.replace('angle', 'Ang')
        print(f"  {clean:<20} {coef:>10.3f} {se:>8.3f} "
              f"{t:>8.3f} {p:>10.4f}  {sig}")

# ============================================================
# STEP 5 — PEAK 1 vs PEAK 2 COMPARISON
# ============================================================
print("\n\n" + "="*60)
print("COMPARISON — Peak 1 vs Peak 2")
print("="*60)
print(f"\n  {'Target':<15} {'Peak':<10} {'R²':>6} {'RMSE (%)':>10}")
print("  " + "-"*45)
for target in ['kneeFres', 'knee_PFc']:
    r1 = pd.DataFrame(results_summary)[
        pd.DataFrame(results_summary)['Target']==target].iloc[0]
    r2 = pd.DataFrame(results_summary_p2)[
        pd.DataFrame(results_summary_p2)['Target']==target].iloc[0]
    flag = '⚠' if r2['R²'] < 0.80 else '✓'
    print(f"  {target:<15} {'Peak 1':<10} {r1['R²']:>6.3f} "
          f"{r1['RMSE (%)']:>9.1f}%  ✓")
    print(f"  {target:<15} {'Peak 2':<10} {r2['R²']:>6.3f} "
          f"{r2['RMSE (%)']:>9.1f}%  {flag}")
    print()
```

```python
# ============================================================
# EXTRACT P1 AND P2 USING GRF_vert AS REFERENCE
# All other variables are extracted AT THE SAME TIME POINT
# ============================================================

stance_pct = np.linspace(0, 100, 101)

def interp_stance(arr, stance_fraction):
    arr = np.array(arr, dtype=float)
    n   = len(arr)
    if n < 2:
        return np.full(101, np.nan)
    n_stance = max(2, round(stance_fraction * n))
    arr_s    = arr[:n_stance]
    x_orig   = np.linspace(0, 100, len(arr_s))
    f = interp1d(x_orig, arr_s, kind='linear',
                 bounds_error=False, fill_value=np.nan)
    return f(stance_pct)


def find_grf_peaks(grf_curve):
    """
    Find P1 and P2 indices in a normalized GRF_vert curve.
    P1 = max in 5–40%
    Valley = min between P1 and 85%
    P2 = max between valley and 90%
    Returns (idx_p1, idx_p2) as indices into the 101-point array.
    """
    # P1
    m1   = (stance_pct >= 5) & (stance_pct <= 40)
    seg1 = grf_curve[m1]
    if np.all(np.isnan(seg1)):
        return None, None
    idx_p1_local = np.nanargmax(seg1)
    idx_p1       = np.where(m1)[0][idx_p1_local]
    pct_p1       = stance_pct[idx_p1]

    # Valley: min after P1 up to 80%
    m_v   = (stance_pct > pct_p1) & (stance_pct <= 80)
    seg_v = grf_curve[m_v]
    if np.all(np.isnan(seg_v)):
        return idx_p1, None
    idx_v_local = np.nanargmin(seg_v)
    idx_v       = np.where(m_v)[0][idx_v_local]
    pct_v       = stance_pct[idx_v]

    # P2: max after valley up to 90%
    m2   = (stance_pct > pct_v) & (stance_pct <= 90)
    seg2 = grf_curve[m2]
    if np.all(np.isnan(seg2)) or len(seg2) == 0:
        return idx_p1, None
    idx_p2_local = np.nanargmax(seg2)
    idx_p2       = np.where(m2)[0][idx_p2_local]

    return idx_p1, idx_p2


# Variables to extract (all time series)
ts_vars   = ['GRF_vert', 'GRF_ap', 'GRF_ml',
             'kneeFres', 'knee_PFc', 'ankleFres', 'hipFres',
             'angleKnee', 'angleHip', 'angleAnkle',
             'momentKnee', 'momentHip', 'momentAnkle']

meta_cols = ['id', 'file', 'side', 'slope', 'condition',
             'height', 'weight', 'age', 'speed',
             'step_length', 'stride_length', 'gc_dur',
             'stance', 'cadence']

rows_p1 = []
rows_p2 = []
skipped = 0

for _, row in df_merged.iterrows():
    sf   = row['stance']
    meta = {col: row[col] for col in meta_cols if col in row.index}

    # --- Get GRF_vert curve to find peak indices ---
    grf_raw = row.get('GRF_vert', None)
    if not isinstance(grf_raw, (list, np.ndarray)) or len(grf_raw) < 2:
        skipped += 1
        continue

    grf_curve = interp_stance(grf_raw, sf)
    idx_p1, idx_p2 = find_grf_peaks(grf_curve)

    if idx_p1 is None:
        skipped += 1
        continue

    entry_p1 = meta.copy()
    entry_p1['grf_pct_p1'] = stance_pct[idx_p1]

    entry_p2 = meta.copy()
    entry_p2['grf_pct_p2'] = stance_pct[idx_p2] if idx_p2 is not None else np.nan

    # --- Extract all variables AT the peak indices ---
    for var in ts_vars:
        raw = row.get(var, None)
        if not isinstance(raw, (list, np.ndarray)) or len(raw) < 2:
            entry_p1[var] = np.nan
            entry_p2[var] = np.nan
            continue

        curve = interp_stance(raw, sf)
        entry_p1[var] = curve[idx_p1] if idx_p1 is not None else np.nan
        entry_p2[var] = (curve[idx_p2] if idx_p2 is not None
                         else np.nan)

    rows_p1.append(entry_p1)
    rows_p2.append(entry_p2)

df_p1_final = pd.DataFrame(rows_p1)
df_p2_final = pd.DataFrame(rows_p2)

for df_ in [df_p1_final, df_p2_final]:
    df_['id']    = df_['id'].astype(str).str.strip()
    df_['slope'] = df_['slope'].astype(int)

# Exclude 16, 26, 20
exclude = ['16', '26', '20']
df_p1_final = df_p1_final[~df_p1_final['id'].isin(exclude)].copy()
df_p2_final = df_p2_final[~df_p2_final['id'].isin(exclude)].copy()

# Normalize forces to ×BW
force_cols = ['GRF_vert', 'GRF_ap', 'GRF_ml',
              'kneeFres', 'knee_PFc', 'ankleFres', 'hipFres']
for df_ in [df_p1_final, df_p2_final]:
    for col in force_cols:
        if col in df_.columns:
            df_[col] = df_[col] / (df_['weight'] * 9.81)

print("✓ GRF-based peak extraction complete")
print(f"  Skipped rows (no GRF data): {skipped}")
print(f"  P1 rows: {len(df_p1_final)} | "
      f"Participants: {df_p1_final['id'].nunique()}")
print(f"  P2 rows: {len(df_p2_final)} | "
      f"Participants: {df_p2_final['id'].nunique()}")

# --- Verify P2 timing ---
print(f"\n  Mean P2 timing (% stance) per slope:")
for slope in [0, -6, -12, -18]:
    v = df_p2_final[df_p2_final['slope']==slope]['grf_pct_p2'].dropna()
    print(f"    {slope:>4}°: {v.mean():.1f} ± {v.std():.1f}%  "
          f"(n={len(v)})")

# --- Sanity check values ---
print(f"\n  PEAK 1 sanity check (0°) — should match Alexander Table 2:")
for var, ref in [('kneeFres', '4.09'), ('knee_PFc', '0.86'),
                 ('GRF_vert', '~1.2')]:
    v = df_p1_final[df_p1_final['slope']==0][var].dropna()
    print(f"    {var:<15}: {v.mean():.3f} ± {v.std():.3f} ×BW  "
          f"(Alexander: {ref} ×BW)")

print(f"\n  PEAK 2 sanity check (0°) — should match Alexander Table 2:")
for var, ref in [('kneeFres', '4.09'), ('knee_PFc', '0.86'),
                 ('GRF_vert', '~1.2')]:
    v = df_p2_final[df_p2_final['slope']==0][var].dropna()
    print(f"    {var:<15}: {v.mean():.3f} ± {v.std():.3f} ×BW  "
          f"(Alexander: {ref} ×BW)")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# ============================================================
# COMPUTE TIME DELAY BETWEEN GRF PEAK AND knee_PFc PEAK
# For Peak 1 and Peak 2
# ============================================================

stance_pct = np.linspace(0, 100, 101)

def interp_stance(arr, stance_fraction):
    arr = np.array(arr, dtype=float)
    n   = len(arr)
    if n < 2:
        return np.full(101, np.nan)
    n_stance = max(2, round(stance_fraction * n))
    arr_s    = arr[:n_stance]
    x_orig   = np.linspace(0, 100, len(arr_s))
    f = interp1d(x_orig, arr_s, kind='linear',
                 bounds_error=False, fill_value=np.nan)
    return f(stance_pct)


def find_grf_peaks(grf_curve):
    """Returns (idx_p1, idx_p2) in the 101-point stance array."""
    # P1: max in 5–40%
    m1 = (stance_pct >= 5) & (stance_pct <= 40)
    s1 = grf_curve[m1]
    if np.all(np.isnan(s1)):
        return None, None
    idx_p1 = np.where(m1)[0][np.nanargmax(s1)]
    pct_p1 = stance_pct[idx_p1]

    # Valley: min after P1 up to 80%
    mv  = (stance_pct > pct_p1) & (stance_pct <= 80)
    sv  = grf_curve[mv]
    if np.all(np.isnan(sv)):
        return idx_p1, None
    idx_v   = np.where(mv)[0][np.nanargmin(sv)]
    pct_v   = stance_pct[idx_v]

    # P2: max after valley up to 90%
    m2 = (stance_pct > pct_v) & (stance_pct <= 90)
    s2 = grf_curve[m2]
    if np.all(np.isnan(s2)) or len(s2) == 0:
        return idx_p1, None
    idx_p2 = np.where(m2)[0][np.nanargmax(s2)]
    return idx_p1, idx_p2


def find_pfc_peak_near(pfc_curve, ref_pct, window=15):
    """Find PFc peak within ±window% of the reference GRF peak."""
    lo = max(0, ref_pct - window)
    hi = min(100, ref_pct + window)
    mask = (stance_pct >= lo) & (stance_pct <= hi)
    seg  = pfc_curve[mask]
    sp   = stance_pct[mask]
    if np.all(np.isnan(seg)) or len(seg) == 0:
        return np.nan
    return sp[np.nanargmax(seg)]


# --------------------------------------------------------
# Main loop: compute delays per trial
# --------------------------------------------------------
rows_delay = []
exclude = ['16', '20', '26']

for _, row in df_merged.iterrows():
    pid   = str(row['id']).strip()
    if pid in exclude:
        continue

    slope = int(row['slope'])
    sf    = row['stance']
    gc    = row['gc_dur']   # gait cycle duration in seconds

    grf_raw = row.get('GRF_vert', None)
    pfc_raw = row.get('knee_PFc', None)

    if (not isinstance(grf_raw, (list, np.ndarray)) or len(grf_raw) < 2 or
        not isinstance(pfc_raw, (list, np.ndarray)) or len(pfc_raw) < 2):
        continue

    grf_curve = interp_stance(grf_raw, sf)
    pfc_curve = interp_stance(pfc_raw, sf)

    idx_p1, idx_p2 = find_grf_peaks(grf_curve)

    if idx_p1 is None:
        continue

    # Stance duration in seconds = gc_dur * stance_fraction
    stance_dur = gc * sf   # seconds for full stance

    # --- Peak 1 ---
    grf_p1_pct = stance_pct[idx_p1]
    pfc_p1_pct = find_pfc_peak_near(pfc_curve, grf_p1_pct, window=15)
    if not np.isnan(pfc_p1_pct):
        # Convert % stance difference to seconds
        delay_p1_s = abs(pfc_p1_pct - grf_p1_pct) / 100 * stance_dur
        rows_delay.append({
            'id': pid, 'slope': slope, 'peak': 'Peak 1',
            'grf_pct': grf_p1_pct, 'pfc_pct': pfc_p1_pct,
            'delay_s': delay_p1_s,
        })

    # --- Peak 2 ---
    if idx_p2 is not None:
        grf_p2_pct = stance_pct[idx_p2]
        pfc_p2_pct = find_pfc_peak_near(pfc_curve, grf_p2_pct, window=15)
        if not np.isnan(pfc_p2_pct):
            delay_p2_s = abs(pfc_p2_pct - grf_p2_pct) / 100 * stance_dur
            rows_delay.append({
                'id': pid, 'slope': slope, 'peak': 'Peak 2',
                'grf_pct': grf_p2_pct, 'pfc_pct': pfc_p2_pct,
                'delay_s': delay_p2_s,
            })

df_delay = pd.DataFrame(rows_delay)

print(f"✓ Delay computed: {len(df_delay)} rows")
print(f"\n  Mean delay (s) per slope and peak:")
summary = (df_delay.groupby(['peak', 'slope'])['delay_s']
           .agg(['mean', 'std', 'count'])
           .round(4))
print(summary)

# ============================================================
# PLOT — one panel per peak, same figure
# ============================================================

slopes_ordered = [0, -6, -12, -18]
colors_slope   = {0: '#a8d5e2', -6: '#5b9fbf', -12: '#2a6a96', -18: '#0d2d55'}
slope_labels   = {0: '0°', -6: '−6°', -12: '−12°', -18: '−18°'}

fig, axes = plt.subplots(1, 2, figsize=(13, 5.5), sharey=False)
fig.suptitle('Absolute Time Delay between GRF Peak and PFc Peak\n'
             'by Slope and Phase | Excluded: [16, 20, 26]',
             fontsize=13, fontweight='bold')

for ax, peak_label in zip(axes, ['Peak 1', 'Peak 2']):

    df_pk = df_delay[df_delay['peak'] == peak_label]

    means = []
    stds  = []
    cols  = []
    xlbls = []

    for slope in slopes_ordered:
        # Peak 2 at slope 0 may be unreliable — skip if few data
        df_s = df_pk[df_pk['slope'] == slope]['delay_s'].dropna()
        if len(df_s) < 3:
            continue
        means.append(df_s.mean())
        stds.append(df_s.std())
        cols.append(colors_slope[slope])
        xlbls.append(slope_labels[slope])

    x = np.arange(len(xlbls))

    bars = ax.bar(x, means, yerr=stds,
                  color=cols, width=0.55,
                  error_kw=dict(ecolor='#444444', elinewidth=1.2,
                                capsize=5, capthick=1.2),
                  edgecolor='white', linewidth=0.5,
                  zorder=3)

    # Value labels on top of bars
    for bar, m, s in zip(bars, means, stds):
        ax.text(bar.get_x() + bar.get_width()/2,
                m + s + 0.0005,
                f'{m:.3f}s',
                ha='center', va='bottom',
                fontsize=8.5, color='#333333')

    ax.set_xticks(x)
    ax.set_xticklabels(xlbls, fontsize=10)
    ax.set_xlabel('Slope (°)', fontsize=11, labelpad=5)
    ax.set_ylabel('Absolute Time Delay (s)', fontsize=11, labelpad=5)
    ax.set_title(f'{peak_label}', fontsize=12, fontweight='bold')
    ax.set_ylim(0, max(means) + max(stds) * 2 + 0.005)
    ax.grid(True, axis='y', linestyle=':', alpha=0.4, color='#aaaaaa')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.tick_params(labelsize=9)

    # Annotation box
    ax.text(0.97, 0.97,
            'Tight synchronization\nbetween GRF and PFc peaks\njustifies GRF as temporal trigger',
            transform=ax.transAxes, ha='right', va='top',
            fontsize=7.5, style='italic', color='#2a6a96',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='#e8f4f8',
                      edgecolor='#2a6a96', alpha=0.9))

plt.tight_layout()
fig.savefig("figures/time_delay_grf_pfc_p1_p2.png",
            dpi=300, bbox_inches='tight', facecolor='white')
print("\n✓ Saved: figures/time_delay_grf_pfc_p1_p2.png")
plt.show()

# ============================================================
# PRINT SUMMARY TABLE
# ============================================================
print("\n" + "="*65)
print("TIME DELAY SUMMARY — Mean ± SD (seconds)")
print("="*65)
print(f"\n  {'Peak':<10} {'Slope':>6} {'Mean (s)':>10} "
      f"{'SD (s)':>10} {'n':>5}")
print("  " + "-"*45)

for peak in ['Peak 1', 'Peak 2']:
    for slope in slopes_ordered:
        d = df_delay[(df_delay['peak']==peak) &
                     (df_delay['slope']==slope)]['delay_s'].dropna()
        if len(d) < 3:
            continue
        print(f"  {peak:<10} {slope:>6}°  "
              f"{d.mean():>10.4f}  {d.std():>10.4f}  {len(d):>5}")
    print()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import statsmodels.formula.api as smf

# ============================================================
# CONFIGURATION
# ============================================================
EXCLUDE_IDS = ['16', '20', '23']  # ← añadidos 16 y 23

df_p1 = df_peak1[~df_peak1['id'].astype(str).str.strip().isin(EXCLUDE_IDS)].copy()
df_p1['id']    = df_p1['id'].astype(str).str.strip()
df_p1['slope'] = df_p1['slope'].astype(int)

# ← CORRECTED per Feature Selection (rows 0 & 6 from the table)
model_config = {
    'kneeFres': {
        'features': ['angleKnee', 'angleHip', 'angleAnkle',   # ← angleAnkle added back
                     'height', 'weight', 'speed', 'slope'],   #   cadence excluded
        'label':    'Tibiofemoral Joint Compression Force',
        'scenario': 'No Cadence (Row 0)',
    },
    'knee_PFc': {
        'features': ['angleKnee', 'angleHip',                  # ← angleAnkle removed
                     'height', 'weight', 'speed', 'slope'],   #   cadence & ankle excluded
        'label':    'Patellofemoral Joint Compression Force',
        'scenario': 'No Cadence & No Ankle (Row 6)',
    },
}

slopes_ordered = [0, -6, -12, -18]
colors_slope   = {0: '#a8d5e2', -6: '#5b9fbf', -12: '#2a6a96', -18: '#0d2d55'}

# ============================================================
# FIT + PLOT
# ============================================================
results_summary = []

fig, axes = plt.subplots(1, 2, figsize=(16, 7))
fig.suptitle('Linear Regression — Peak 1 (Absorption Phase)\n'
             'Subject ID as Fixed Covariate | Participants 16, 20 & 23 Excluded',
             fontsize=13, fontweight='bold')

for ax, (target, cfg) in zip(axes, model_config.items()):

    features = cfg['features']
    label    = cfg['label']

    all_cols = list(dict.fromkeys(['id', 'slope', target] + features))
    df_c     = df_p1[all_cols].dropna().copy().reset_index(drop=True)
    df_c['slope'] = df_c['slope'].astype(int)

    # --- Build X matrix ---
    subj_dummies = pd.get_dummies(df_c['id'], prefix='Subj', drop_first=True)
    X = pd.concat([df_c[features].reset_index(drop=True),
                   subj_dummies.reset_index(drop=True)], axis=1)
    y          = df_c[target].values.astype(float)
    slope_arr  = df_c['slope'].values.astype(int)

    # --- Fit ---
    model  = LinearRegression()
    model.fit(X.values, y)
    y_pred = model.predict(X.values)

    # --- Stats ---
    r2       = r2_score(y, y_pred)
    rmse     = np.sqrt(mean_squared_error(y, y_pred))
    rmse_pct = (rmse / y.mean()) * 100
    n        = len(y)

    # --- Equation string ---
    eq_lines = [f"y = {model.intercept_:.2f}"]
    for name, coef in zip(X.columns, model.coef_):
        if 'Subj' not in name:
            clean = name.replace('angle', 'Ang')
            eq_lines.append(f"{coef:+.2f} × {clean}")
    eq_lines.append("[+ Subject Effects]")
    eq_str = "\n".join(eq_lines)

    results_summary.append({
        'Target':   target,
        'Scenario': cfg['scenario'],
        'n':        n,
        'R²':       round(r2, 3),
        'RMSE (%)': round(rmse_pct, 2),
        'Features': ", ".join(features),
    })

    # --- Scatter ---
    for slope in slopes_ordered:
        idx = np.where(slope_arr == slope)[0]
        if len(idx) == 0:
            continue
        ax.scatter(y[idx], y_pred[idx],
                   color=colors_slope[slope], alpha=0.7, s=55,
                   label=f'{slope}°', edgecolors='white',
                   linewidths=0.4, zorder=3)

    # Perfect fit line
    vmin = min(y.min(), y_pred.min())
    vmax = max(y.max(), y_pred.max())
    pad  = (vmax - vmin) * 0.05
    ax.plot([vmin - pad, vmax + pad],
            [vmin - pad, vmax + pad],
            color='red', linestyle='--', linewidth=1.8,
            label='Perfect Fit', zorder=2)

    # Stats box
    stats_text = (f"n = {n}\n"
                  f"$R^2$ = {r2:.3f}\n"
                  f"RMSE = {rmse_pct:.1f}%\n\n"
                  f"{eq_str}")
    ax.text(0.03, 0.97, stats_text,
            transform=ax.transAxes, fontsize=7.5,
            verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='white',
                      alpha=0.9, edgecolor='#cccccc'))

    ax.set_xlabel(f"Actual {label}", fontsize=10)
    ax.set_ylabel(f"Predicted {label}", fontsize=10)
    ax.set_title(f"Target: {label}", fontsize=11, fontweight='bold')
    ax.set_xlim(vmin - pad, vmax + pad)
    ax.set_ylim(vmin - pad, vmax + pad)
    ax.grid(True, linestyle=':', alpha=0.4)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.tick_params(labelsize=9)

    handles, lbls = ax.get_legend_handles_labels()
    ax.legend(handles, lbls, title='Slope (°)', fontsize=8,
              title_fontsize=8.5, loc='lower right',
              framealpha=0.9, edgecolor='#cccccc')

plt.tight_layout()
fig.savefig("figures/regression_peak1_final.png",
            dpi=300, bbox_inches='tight', facecolor='white')
plt.show()

# ============================================================
# SUMMARY TABLE
# ============================================================
print(pd.DataFrame(results_summary).to_string(index=False))

# ============================================================
# COEFFICIENTS WITH p-VALUES (statsmodels)
# ============================================================
for target, cfg in model_config.items():
    features = cfg['features']
    label    = cfg['label']

    all_cols = list(dict.fromkeys(['id', 'slope', target] + features))
    df_c     = df_p1[all_cols].dropna().copy().reset_index(drop=True)
    df_c['slope'] = df_c['slope'].astype(int)
    df_c['id']    = df_c['id'].astype(str)

    formula  = f"{target} ~ " + " + ".join(features) + " + C(id)"
    sm_model = smf.ols(formula, data=df_c).fit()

    print(f"\n[{label}]")
    print(f"R² = {sm_model.rsquared:.3f} | Adj. R² = {sm_model.rsquared_adj:.3f} | n = {int(sm_model.nobs)}")
    print(f"\n  {'Variable':<20} {'Coef':>10} {'SE':>8} {'t':>8} {'p':>10}  Sig")
    print("  " + "-"*65)

    for name in sm_model.params.index:
        if 'C(id)' in name:
            continue
        coef = sm_model.params[name]
        se   = sm_model.bse[name]
        t    = sm_model.tvalues[name]
        p    = sm_model.pvalues[name]
        sig  = ('***' if p < 0.001 else '**' if p < 0.01 else '*' if p < 0.05 else 'ns')
        clean = name.replace('angle', 'Ang')
        print(f"  {clean:<20} {coef:>10.3f} {se:>8.3f} {t:>8.3f} {p:>10.4f}  {sig}")
```
